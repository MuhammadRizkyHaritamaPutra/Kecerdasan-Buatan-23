```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#membaca dataset
df = pd.read_csv("HepatitisCdata.csv")
```

1.Tampilkan rata-rata, standar deviasi, nilai minimum, nilai maksimum, Q1, Q2, dan Q3 dari attribute bertipe angka.

```python
# Memilih hanya kolom numerik
numeric_df = df.select_dtypes(include=['int64', 'float64'])
# Menghitung statistik deskriptif
stats = numeric_df.describe(percentiles=[0.25, 0.5, 0.75]).T
# Merapikan nama kolom
stats = stats.rename(columns={
    'mean': 'Rata-rata',
    'std': 'Standar Deviasi',
    'min': 'Minimum',
    '25%': 'Q1',
    '50%': 'Q2 (Median)',
    '75%': 'Q3',
    'max': 'Maksimum'
})
# Menampilkan hanya kolom yang dibutuhkan
stats
```

3. Tampilkan informasi jumlah record, jumlah attribute, nama attribute, jumlah attribute dengan tipe angka.

```python
# Jumlah record (baris)
jumlah_record = df.shape[0]

# Jumlah attribute (kolom)
jumlah_attribute = df.shape[1]

# Nama-nama attribute
nama_attribute = df.columns.tolist()

# Jumlah attribute dengan tipe angka
jumlah_atribut_numerik = df.select_dtypes(include=['int64', 'float64']).shape[1]

# Menampilkan hasil
print("Jumlah record (baris):", jumlah_record)
print("Jumlah attribute (kolom):", jumlah_attribute)
print("Nama-nama attribute:")
print(nama_attribute)
print("Jumlah attribute bertipe angka:", jumlah_atribut_numerik)
```

4. Tampilkan jumlah label dari semua attribute yang bernilai object

```python
# Memilih atribut bertipe object (kategori)
obj_columns = df.select_dtypes(include=['object'])

# Menampilkan jumlah data & jumlah label unik
print("Informasi atribut bertipe object:\n")
for col in obj_columns.columns:
    total_data = obj_columns[col].count()
    unique_labels = obj_columns[col].nunique()

    print(f"- {col}:")
    print(f"   Jumlah data      : {total_data}")
    print(f"   Jumlah label unik: {unique_labels}")
```

5. Visualisasikan dataset dari kolom category

```python
# Hitung jumlah data untuk tiap kategori
category_counts = df['Category'].value_counts()

# Membuat diagram batang
plt.figure(figsize=(8,6))
bars = category_counts.plot(kind='bar', color='skyblue', edgecolor='black')

# Menambahkan label jumlah di atas batang
for i, count in enumerate(category_counts):
    plt.text(i, count + 1, str(count), ha='center', fontsize=10)

#  Memberi judul dan label
plt.title('Distribusi Jumlah Data per Category', fontsize=14)
plt.xlabel('Category', fontsize=12)
plt.ylabel('Jumlah Data', fontsize=12)
plt.xticks(rotation=45)
plt.grid(axis='y', linestyle='--', alpha=0.7)

# Tampilkan grafik
plt.tight_layout()
plt.show()
```

6. Tampilkan korelasi dataset menggunakan heatmap.

```python
# Memilih hanya atribut numerik
numeric_df = df.select_dtypes(include=['int64', 'float64'])

# Menghitung matriks korelasi
corr_matrix = numeric_df.corr()

# Membuat heatmap
plt.figure(figsize=(12,8))
sns.heatmap(corr_matrix,
            annot=True,
            fmt=".2f",
            cmap='coolwarm',
            cbar=True)

# Judul grafik
plt.title('Heatmap Korelasi Atribut Numerik', fontsize=16)

# Tampilkan grafik
plt.tight_layout()
plt.show()
```
